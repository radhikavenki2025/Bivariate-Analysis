```python
#####Test the similarity between the degree_t(Sci&Tech) and specialisation(Mkt&HR) with respect to salary at significance level of 5%.
#####(Make decision using Hypothesis Testing)
```

### One Sample t-test

### Salary - degree_t ( Sci&tech )
### Salary - specialisation ( Mkt&HR )


### Null Hypothesis : The mean salary of Sci & Tech students with Mkt/HR specialisation is equal to the overall mean salary.

### Alternate Hypothesis : The mean salary of Sci & Tech students with Mkt/HR specialisation is different from the overall mean salary.

### Significance level : α=0.05

```python
from scipy.stats import ttest_1samp

# Subset: Sci & Tech with Mkt + HR
subset = df_new[
            (df_new['degree_t'] == 'Sci&Tech') &
            (df_new['specialisation'].isin(['Mkt&HR']))
               ]['salary'].dropna()

# Benchmark: overall mean salary
overall_mean = df_new['salary'].mean()

# One-sample t-test
t_stat, p_value = ttest_1samp(subset, overall_mean)

print("t-statistic:", t_stat)
print("p-value:", p_value)
```

#### Result : p-value > 0.05 ; At the 5% significance level, there is no statistically significant difference between the mean salary of Sci & Tech students specialising in Marketing and HR and the overall mean salary.
### Hence, salaries are similar.

#### Test the similarity between the degree_t(Sci&Tech)with respect to
#### etest_p and mba_p at significance level of 5%.(Make decision using Hypothesis Testing)

### Null Hypothesis : There is no significant difference in etest_p & mba_p test scores of Sci&Tech degree students.
### Alternate Hypothesis : There is significant difference in etest & mba_p test scores of Sci&Tech degree students.

```python
from scipy.stats import ttest_rel

df_new=df_new.dropna()

etest_p = df_new.loc[df_new['degree_t'] == 'Sci&Tech', 'etest_p']

mba_p = df_new.loc[df_new['degree_t']=='Sci&Tech','mba_p']


t_stat,p_value = ttest_rel(etest_p,mba_p)


print("t-statistic:", t_stat)
print("p-value:", p_value)
```

### Results : p-value <0.05 implies that reject the Null Hypothesis , Hence there is significant difference in 
### etest_p & mba_p scores of Sci&Tech degree students.In other words Sci& Tech students have significant difference in their etest_p & mba_p test scores.

#### Probability Density Function 

```python
def get_pdf_probability(df_new,startrange,endrange):
    from matplotlib import pyplot
    from scipy.stats import norm
    import seaborn as sns
    ax=sns.distplot(df_new,kde=True,kde_kws={'color':'blue'},color='Green' )
    
    pyplot.axvline(startrange,color='Red')
    pyplot.axvline(endrange,color='Red')

    sample=df_new
    sample_mean=sample.mean()
    sample_std=sample.std()

    print('Mean=%.3f, Standard Deviation=%.3f' % (sample_mean,sample_std))

    dist=norm(sample_mean , sample_std)

    values=[ value for value in range (startrange,endrange)]

    probabilities= [ dist.pdf(value) for value in values ]

    prob = sum (probabilities)

    print("The area between {},{} : {}".format(startrange,endrange,sum(probabilities)))

    return prob
```

```python
get_pdf_probability(df_new["salary"],700000,900000)
```
