Independent Sample - Unpaired T-Test

Group : Different ( Male / Female ) ; Condition : Same ( Salary ) Different Group - Same Condition - Unpaired

```python
from scipy.stats import ttest_ind

dataset=dataset.dropna()

####print(dataset.columns)

male = dataset.loc[dataset['gender'] == 'M', 'salary']

female=dataset.loc[dataset['gender']=='F','salary']

ttest_ind(male,female)

```

```python
##### Independent / Unpaired 
##### Group - ssc_b : Central / Others ; Condition - Salary


from scipy.stats import ttest_ind

dataset=dataset.dropna()

####print(dataset.columns)

Central = dataset.loc[dataset['ssc_b'] == 'Central', 'salary']

Others =dataset.loc[dataset['ssc_b']=='Others','salary']

ttest_ind(Central,Others)

```

```python
##### Independent / Unpaired 
##### Group - hscc_b : Central / Others ; Condition - Salary


from scipy.stats import ttest_ind

dataset=dataset.dropna()

####print(dataset.columns)

Central = dataset.loc[dataset['hsc_b'] == 'Central', 'salary']

Others =dataset.loc[dataset['hsc_b']=='Others','salary']

ttest_ind(Central,Others)
```
