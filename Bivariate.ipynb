```python
##### Variance Inflation factor
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

def calc_vif(X):
    ## calculating VIF
    vif=pd.DataFrame()
    vif["variables"] = X.columns
    vif["VIF"] = [variance_inflation_factor(X.values,i) 
                  for i in range (X.shape[1])]

    return(vif)
                  
    
```

```python

#### to call quantitative columns only

calc_vif(dataset[["ssc_p","hsc_p","degree_p","etest_p","mba_p","salary"]])
```

```python
calc_vif(dataset[['ssc_p','hsc_p','etest_p','salary']])
```

Independent Sample - Unpaired T-Test

Group : Different ( Male / Female ) ; Condition : Same ( Salary ) 
Different Group - Same Condition - Unpaired

```python
from scipy.stats import ttest_ind

dataset=dataset.dropna()

####print(dataset.columns)

male = dataset.loc[dataset['gender'] == 'M', 'salary']

female=dataset.loc[dataset['gender']=='F','salary']

ttest_ind(male,female)

```

##### p value >0.05 Not statistically significant difference

```python
from scipy.stats import ttest_rel


clean_data = dataset.dropna(subset=['hsc_s', 'mba_p', 'etest_p'])


mba = clean_data.loc[clean_data['hsc_s'] == 'Commerce', 'mba_p']
etest = clean_data.loc[clean_data['hsc_s'] == 'Commerce', 'etest_p']


result = ttest_rel(mba.values, etest.values)

print(result)


```

##### p value >0.05 Not statistically significant difference

Reject NUll Hypothesis p<0.05

Accept NUll Hypothesis and Reject Alternate Hypothesis

ANAVO : Analysis of Variance


```python
#### One way Classification 

import scipy.stats as stats

stats.f_oneway(dataset['ssc_p'],dataset['hsc_p'],dataset['degree_p'])
```

##### Statement : Difference of Variances between SSC / HSC / Degree

#### The above p value > 0.05 hence we reject the null hypothesis.

##### Null hypothesis : There is no significant difference between the columns

##### Alternate hypothesis : There is significant difference between the columns


```python
import statsmodels.api as sm

from statsmodels.formula.api import ols

clean_data = dataset.dropna(subset=['gender', 'specialisation', 'salary'])

model = ols('salary ~ gender + specialisation', data=clean_data).fit()

anova_table = sm.stats.anova_lm(model, typ=2)
print(anova_table)
```
